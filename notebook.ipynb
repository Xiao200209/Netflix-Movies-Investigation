**What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration.**

```python
# Importing pandas and matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
netflix_df.head(5)
```

```python
selected_year = netflix_df[(netflix_df['release_year'] >= 1990) 
                           & (netflix_df['release_year'] <= 1999) 
                           & (netflix_df['type']=='Movie')]
```

```python
plt.hist(x = selected_year['duration'] , bins = 30)
plt.xlabel("Duration")
plt.ylabel("Number of Movies")
plt.title("Histogram of Movie Duration")
plt.show()
```

**The most frequent duration is around 100 mins.**

**A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count.**

```python
short_movie_count = selected_year[(selected_year['duration']< 90) 
                                  & (selected_year['genre'] == 'Action' ) ]['title'].count()

print('There are '+ str(short_movie_count) + ' short action movies released in the 1990s.')
```
